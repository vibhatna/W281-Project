```python
import os
import cv2
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from progressbar import ProgressBar
```

```python
df1 = pd.read_csv('TargetData-200-Train.csv')
df2 = pd.read_csv('TargetData-Train.csv')
df3 = pd.read_csv('TargetData-TF-Train.csv')
```

```python
def showBalance(df):
    temparray = np.array(df['label'])
    uniq, counts = np.unique(temparray, return_counts=True)
    plt.bar(uniq, counts, 1)
    plt.title('Class Frequency')
    plt.xlabel('Class')
    plt.ylabel('Frequency')
    plt.show()
def showBalance1(df):
    temparray = np.array(df['landmark_id'])
    uniq, counts = np.unique(temparray, return_counts=True)
    plt.bar([1,2,3,4,5], counts, 1)
    plt.title('Class Frequency')
    plt.xlabel('Class')
    plt.ylabel('Frequency')
    plt.show()

```

### 5 classes results

```python
showBalance1(df3)
```

### 51 classes results

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 5))

ax[0].hist(df2['label'],bins = 51, histtype='bar', ec='black')
ax[0].set_xlabel('Landmark Class Label')
ax[0].set_ylabel('# of Images in Training Data')
ax[0].set_title("Frequency of Classes in 51 Class Training Data")

ax[1].hist(df1['landmark_id'],bins = 491, histtype='bar', ec='black')
ax[1].set_xlabel('Landmark Class Label')
ax[1].set_ylabel('# of Images in Training Data')
ax[1].set_title("Frequency of Classes in 491 Class Training Data")

plt.show()
```
